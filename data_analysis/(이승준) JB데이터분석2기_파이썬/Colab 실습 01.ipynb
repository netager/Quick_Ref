```python
## 정리
import pandas as pd
import glob

file_lists = glob.glob('*.xlsx')

df = pd.DataFrame()
for file_list in file_lists:
    col = file_list.replace('.xlsx', '')
    tmp = pd.read_excel(file_list, index_col='Date')
    df[col] = tmp['Close']

df.plot(secondary_y=['JB금융지주','코스피지수'])
```
