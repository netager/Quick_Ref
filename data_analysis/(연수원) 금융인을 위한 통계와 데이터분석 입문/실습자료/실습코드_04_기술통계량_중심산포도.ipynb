```python
import pandas as pd
apt = pd.read_excel('data/아파트_실거래가_서울.xlsx') 
apt.shape # data frame의 행열의 수 
```

```python
apt.columns
```

```python
# 변수명 변경 
apt.rename(columns = {'전용면적(㎡)' : '면적', '거래금액(만원)' : '가격'}, inplace = True)
apt.columns
```

## 자료의 중심 (center) 

```python
# 평균
apt.가격.mean()
```

```python
# 중위수
apt.가격.median()
```

```python
apt['가격'].median() # 위와 동일 
```

```python
apt[['가격', '면적']].mean()
```

```python
apt[['가격', '면적']].median()
```

- '단가' (단위면적 당 가격 = 가격/면적)라는 새로운 변수 추가하고 이 변수를 분석    

```python
apt['단가'] = apt.가격 / apt.면적
```

```python
apt.단가
```

```python
# 기초통계량 
tb = apt.단가.describe()
tb
```

```python
from scipy import stats
stats.trim_mean(apt.단가, 0.2)
```

## 자료의 산포도 (spread)

```python
# 표준편차 
apt.단가.std()
```

```python
tb['std'] # describe의 결과물을 활용하는 방법 
```

```python
# 범위
tb['max'] - tb['min']
```

```python
# 사분위수 범위 
Q1 = tb['25%']
Q3 = tb['75%']
IQR = Q3 - Q1
print('IQR = ', IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("lower : ", lower, "upper : ", upper)
tf = apt.단가.between(lower,upper) 
outliers = apt.loc[~tf, :] 
outliers
```

```python
print("lower : ", lower, "upper : ", upper)
tf = apt.단가.between(lower,upper)
tf
outliers = apt.loc[~tf, :] 
outliers
```

```python
apt[~tf]
```

## 구별로 단가의 중심, 산포도를 계산하고 비교 

```python
apt.groupby('구').단가.describe()
```

```python
apt.groupby('구').단가.median()
```

```python
apt.groupby('구').단가.median().sort_values() 
```

```python
apt.groupby('구').단가.median().sort_values(ascending = False) 
```

```python
apt.groupby('구').단가.std().sort_values() 
```

```python
# 용산구 자료만 추출 
s = apt[apt.구 == '용산구']
s_tb = s.단가.describe()

Q1 = s_tb['25%']
Q3 = s_tb['75%']
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

tf = s.단가.between(lower,upper) 
outliers = s.loc[~tf, :] 
outliers
```
