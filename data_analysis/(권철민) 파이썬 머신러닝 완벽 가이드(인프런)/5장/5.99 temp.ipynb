```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.preprocessing import PolynomialFeatures

n = 300
x = 4 * np.random.rand(n, 1) - 1
y = 1.5 * x**2 + x + 5 + np.random.rand(n, 1)

plt.scatter(x, y, s=3)
print(x)
```

```python

poly = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly.fit_transform(x)

print(poly.get_feature_names())

plt.scatter(x, x_poly[:,1], s=3)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly.fit_transform(x)
print(poly.get_feature_names())
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_poly, y)

xx = np.linspace(-1, 3, 300)
xx_poly = poly.fit_transform(xx.reshape(-1,1))
y_pred = lr.predict(xx_poly)

plt.figure(figsize=(10, 8))
plt.scatter(x, y, s=3)
plt.plot(xx, y_pred, 'r', label= 'pred', lw=3)
plt.legend()
plt.show()
```
