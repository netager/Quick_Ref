* pandas & matpotlib Package

```
import pandas as pd
import matplotlib.pyplot as plt
```

* .read_csv( )

```
DF = pd.read_csv('PII.csv')

DF.head()
```

# seaborn Package

```
import seaborn as sns
```

# I. 선 그래프

* https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```
plt.figure(figsize = (10, 7))
sns.lineplot(DF.index, 
             DF.Height,
             linewidth = 1,
             color = 'r',
             marker = '>',
             linestyle = '--')
plt.title('Line Graph', size = 30)
plt.xlabel('Index', size = 20)
plt.ylabel('Height', size = 20)
plt.grid(True)
plt.show()
```

# II. 막대 그래프

> ## 1) 연속형 - .barplot( )

* https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot

```
plt.figure(figsize = (10, 7))
sns.barplot(data = DF,
            x = DF.index, 
            y = 'Height')
plt.show()
```

> ## 2) 명목형 - .countplot( )

* https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot

```
plt.figure(figsize = (10, 7))
sns.countplot(data = DF,
              x = 'BloodType')
plt.show()
```

```
plt.figure(figsize = (10, 7))
sns.countplot(data = DF,
              x = 'BloodType',
              hue = 'Gender')
plt.show()
```

```
plt.figure(figsize = (7, 10))
sns.countplot(data = DF,
              y = 'BloodType',
              hue = 'Gender')
plt.show()
```

# III. 히스토그램

* https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

```
plt.figure(figsize = (10, 7))
sns.histplot(data = DF,
             x = 'Height',
             bins = 5, 
             alpha = 0.3)
plt.show()
```

# IV. 상자 그래프

* https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot

```
plt.figure(figsize = (10, 7))
sns.boxplot(data = DF,
            x = 'BloodType', 
            y = 'Height',
            order = ['A', 'B', 'O', 'AB'])
plt.show()
```

# V. 산점도

* https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```
plt.figure(figsize = (10, 7))
sns.scatterplot(data = DF,
                x = 'Height', 
                y = 'Weight', 
                s = 50)
plt.show()
```
