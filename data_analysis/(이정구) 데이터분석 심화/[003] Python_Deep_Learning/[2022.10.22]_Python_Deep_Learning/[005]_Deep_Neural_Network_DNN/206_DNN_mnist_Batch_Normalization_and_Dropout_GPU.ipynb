# MNIST - Categorical Classification
> # Batch Normalization & Dropout


```python
import warnings
warnings.filterwarnings('ignore')
```

# I. MNIST Data_Set Load

```python
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

# II. Data Preprocessing

> ## 1) Reshape and Normalization

* reshape

```python
X_train = X_train.reshape((60000, 28 * 28))
X_test = X_test.reshape((10000, 28 * 28))
```

* Normalization

```python
X_train = X_train.astype(float) / 255
X_test = X_test.astype(float) / 255
```

> ## 2) One Hot Encoding

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

# III. MNIST Keras Modeling

> ## 1) Model Define

* 모델 신경망 구조 정의
 * 활성화 함수의 입력값을 정규화 과정을 수행하여 전달
 * Gradient Vanishing 문제 해결 및 더 큰 Learning Rate를 사용 가능

```python
from tensorflow.keras import models
from tensorflow.keras import layers

mnist = models.Sequential()
mnist.add(layers.Dense(512, input_shape=(28 * 28,)))
mnist.add(layers.BatchNormalization())
mnist.add(layers.Activation('relu'))
mnist.add(layers.Dropout(0.5))
mnist.add(layers.Dense(256))
mnist.add(layers.BatchNormalization())
mnist.add(layers.Activation('relu'))
mnist.add(layers.Dropout(0.5))
mnist.add(layers.Dense(10, activation = 'softmax'))
```

* 모델 구조 확인

```python
mnist.summary()
```

> ## 2) Model Compile

* 모델 학습방법 설정

```python
mnist.compile(loss = 'categorical_crossentropy',
              optimizer = 'rmsprop',
              metrics = ['accuracy'])
```

> ## 3) Model Fit

* 약 6분

```python
%%time

Hist_mnist = mnist.fit(X_train, y_train,
                       epochs = 100,
                       batch_size = 128,
                       validation_split = 0.2)
```

> ## 4) 학습 결과 시각화

* Loss Visualization

* Loss & Accuracy

```python
loss, accuracy = mnist.evaluate(X_test, y_test)

print('Loss = {:.5f}'.format(loss))
print('Accuracy = {:.5f}'.format(accuracy))
```
