> ## 2) 데이터 읽어오기

* pandas DataFrame

```python
import pandas as pd

DF = pd.read_csv('../../../DataSets/creditcard.csv')

DF.info()
```

```python
DF.head()
```

* 0 (정상) Class와 1 (사기) Class 개수

```python
DF.Class.value_counts()
```

* 0 (정상) Class와 1 (사기) Class 비율

```python
(DF.Class.value_counts() / DF.shape[0]) * 100
```

> ## 3) Time 열(Column) 삭제

```python
DF.drop('Time', axis = 1, inplace = True)

DF.head(1)
```

> ## 4) train_test_split( )

* X (Input), y (Output) 지정

```python
X = DF.iloc[:,:-1]
y = DF.iloc[:, -1]

X.shape, y.shape
```

* With 'Stratify'

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 2045,
                                                    stratify = y)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

* Train_Data와 Test_Data의 1 (부정) 비율이 균형

```python
print('Train_Data :','\n', (y_train.value_counts() / y_train.shape[0]) * 100)
print('Test_Data :','\n', (y_test.value_counts() / y_test.shape[0]) * 100)
```

# I. MLP Modeling

> ## 1) Train_Data로 모델 생성

* hidden_layer_sizes : 은닉층 노드의 개수
* activation : 활성화 함수
* solver : 최적화 기법
* max_iter : 학습 반복 횟수

```python
from sklearn.neural_network import MLPClassifier

Model_NN = MLPClassifier(hidden_layer_sizes = (11),
                         activation = 'logistic',
                         solver ='adam', 
                         max_iter = 5000,
                         random_state = 2045)

Model_NN.fit(X_train, y_train)
```

> ## 2) Test_Data에 Model 적용

```python
y_hat = Model_NN.predict(X_test)
```

> ## 3) Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_hat)
```

> ## 5) Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_hat, 
                            target_names = ['0', '1'],
                            digits = 5))
```
