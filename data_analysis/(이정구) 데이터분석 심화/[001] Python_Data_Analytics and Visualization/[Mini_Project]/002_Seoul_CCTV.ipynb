# II. 'seoulCCTV.csv' 파일 전처리

> ## 1) 'seoulCCTV.csv' 파일 읽어오기

```python
%matplotlib inline 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd

CCTV = pd.read_csv('../pyData/seoulCCTV.csv')

CCTV.head()
```

# IV. CCTV 데이터와 인구 데이터 합치기

> ## 1) 두 데이터프레임에 공통으로 있는 '구별'로 merge

```python
DF = pd.merge(CCTV, SDFP, on = '구별')
```

```python
DF.head()
```

```python
DF.info()
```

> ## 2) 불필요한 열(Column) 삭제

```python
del DF['2013년도 이전']
del DF['2014년']
del DF['2015년']
del DF['2016년']
```

```python
DF.head()
```

> ## 3) '구별'을 index로 지정

```python
DF.set_index('구별', inplace = True)
```

```python
DF.head()
```

> ## 4) 상관계수(Correlation Coefficient)

* 범위 : -1 ~ 1(0이면 관계없음)
* '고령자비율' vs. '소계'

```python
import numpy as np

print(np.corrcoef(DF['고령자비율'], DF['소계']))
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(DF['고령자비율'], DF['소계'],
            marker='*',
            s = 100,
            c = '#d62728')
plt.show()
```

* '외국인비율' vs. '소계'

```python
print(np.corrcoef()DF['외국인비율'], DF['소계'])
```

* '인구수'     vs. '소계'

```python
plt.figure(figsize = (10, 7))
plt.scatter(DF['외국인비율'], DF['소계'],
            marker='*',
            s = 100,
            c = '#d62728')
plt.show()
```

```python
print(np.corrcoef(DF['인구수'], DF['소계']))
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(DF['인구수'], DF['소계'],
            marker='*',
            s = 100,
            c = '#d62728')
plt.show()
```

> ## 5) CCTV 개수('소계')와 '인구수'의 관계

* '소계'(CCTV 개수) 내림차순 정렬

```python
DF.sort_values(by = '소계', ascending = False).head()
```

* '인구수' 내림차순 정렬

```python
DF.sort_values(by='인구수', ascending=False).head()
```

# V. Visualization

> ## 1) 막대(bar) 그래프 - pandas

* '소계'

```python
import matplotlib.pyplot as plt

DF['소계'].plot(kind = 'barh', grid = True, figsize = (10, 10))
plt.show()
```

* 정렬 - sort_value( )


```python
DF['소계'].sort_values().plot(kind = 'barh', grid = True, figsize = (10, 10))
plt.show()
```

* '인구수' 대비 CCTV비율 계산 후 정렬하여 시각화

```python
DF['CCTV비율'] = DF['소계'] / DF['인구수'] * 100
```

```python
DF['CCTV비율'].sort_values().plot(kind = 'barh', grid = True, figsize = (10, 10))
plt.show()
```

> ## 2) 산점도(scatter) - matplotlib

* '소계' ~ '인구수'

```python
plt.figure(figsize=(10, 7))
plt.scatter(DF['인구수'], DF['소계'], s = 50)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

> ## 3) 회귀선 추가 - seaborn

```python
import seaborn as sns

plt.figure(figsize = (12, 9))

sns.regplot(x = '인구수', y = '소계', data = DF,
            line_kws = {'color':'red'})

for n in range(len(DF)):
    plt.text(DF['인구수'][n] * 1.02, 
             DF['소계'][n] * 0.98, 
             DF.index[n], 
             fontsize = 12)

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```
