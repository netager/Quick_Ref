```python
import pandas as pd

df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0)
df.index = pd.to_datetime(df.index)
df = df.sort_index()[['시가', '저가', '고가', '종가', '거래량']]
df.head()
```

```python
# 인덱스는 기본적으로 그룹화한 날짜 중 마지막 날짜가 선택됩니다. 
# resample 메서드에 다음과 같이 S 옵션을 추가하면 그룹 중 가장 빠른 일자로 인덱스를 설정합니다. 
# 데이터의 가장 빠른 날이 아니라 일반적인 날자의 첫날

df.resample('MS').first()
```

```python
# OHLCV 데이터는 그룹 중 특정 행의 값이 통째로 필요한 것이 아니라 시가는 시작일의 가격, 종가는 마지막 날 가격, 
# 고가는 가장 높은 값, 저가는 가장 낮은 값, 거래량은 합산한 값이 필요합니다. 
# 데이터프레임의 agg 메서드에 컬럼 이름과 적용 방법을 딕셔너리로 기술하면 되겠죠?

how = {'시가':'first', 
       '종가':'last',
       '고가':max,
       '저가':min,
       '거래량':sum}

df.resample('MS').agg(how)  # 월별 통계데이터 생성     
```

```python
# resample 메서드는 월(M)뿐만 아니라 시(H)/분(T)/초(S)/일(D)/주(W) 단위로 데이터를 샘플링할 수 있습니다. 
# M이 "월"로 해석되기 때문에 "분"은 T를 사용하는 것에 주의하세요. 다음 코드는 3일 단위로 데이터를 그룹화합니다.

df.resample('3D').agg(how)
```

```python
# 코드를 실행하면 2021-02-27부터 3일간은 휴일이라 데이터가 존재하지 않아 NaN으로 출력됩니다. 
# 과거 데이터로 백테스팅할 때는 이러한 결측값의 처리를 주의해야 합니다. 
# 데이터프레임의 dropna 메서드를 사용해서 결측값을 제거할 수 있습니다.

df.resample('3D').agg(how).dropna()
```

```python
# resample한 결과에서 대푯값만 변경하고 싶다면 인덱스에 Timedelta 객체로 시간 차이를 보정할 수 있습니다. 
# 다음 코드는 Timestamp 객체로 저장된 데이터프레임의 모든 날짜에 2일의 시간차를 더합니다.

temp = df.resample('3D').agg(how)
temp.index = temp.index + pd.to_timedelta('2D')
temp.dropna()
```

```python
# 위에서 2021-02-15부터 세 개의 데이터를 그룹화하는데, 경우에 따라 2021-02-16부터 그룹화가 필요할 수 있습니다. 
# resample 메서드의 offset 파라미터는 기준 일자를 조정합니다. 
# 다음 코드는 그룹화하는 기준 일자를 1일(1D) 미루기 때문에 2021-02-16에서 2021-02-19까지가 하나의 구간으로 분류됩니다. 
# 이때 2021-02-15는 이전 구간인 2021-02-13부터 2021-02-15구간에 포함됩니다.

df.resample('3D', offset='1D').agg(how)

how = {'시가':'first', 
       '종가':'last',
       '고가':max,
       '저가':min,
       '거래량':sum}
```

```python
# 일봉 데이터를 12시간 단위의 데이터로 업샘플링(up-sampling) 해 보겠습니다.

df.resample('12H').agg(how)
```

```python
# 존재하는 이전 구간의 값을 결측값으로 채워 넣을 수 있습니다. ffill 메서드는 결측값을 이전 로우의 값으로 대체합니다.
# ffill 메서드가 적용된 데이터프레임입니다. 2021-02-15 12:00:00의 행은 이전 위치인 2021-02-15 00:00:00 행의 
# 값으로 채워졌습니다. 참고로 ffill은 forward fill의 약자이며, 비슷한 기능으로 backward fill의 bfill 메서드가 
# 존재합니다. bfill은 아래 행의 값으로 결측값을 채워 넣습니다.

df.resample('12H').agg(how).ffill()
```

```python
# 거래량은 이전의 값이 복사되지 않고 0으로 표기돼 았습니다. 이는 agg 메서드로 거래량 컬럼에 적용한 sum 함수가 결측값을 
# 0으로 합산하기 때문에 발생하는 문제입니다. 제대로 처리하기 위해서는 로우에 하나라도 결측값이 있을 경우 NaN으로 
# 거래량을 설정하는 전처리 과정이 추가돼야 합니다. 드물겠지만 실제 거래량이 없을 수 있어서 행 전체를 비교합니다.

import numpy as np 

temp = df.resample('12H').agg(how)
cond = temp.isna().any(axis=1)
temp.loc[cond, '거래량'] = np.NaN

temp.ffill()
```

## 06) 수익률 계산하기

```python
# 판다스의 데이터프레임은 수익률을 계산하는 pct_change 메서드를 제공합니다. 
# 참고로 pct_change에서 pct는 Percent의 약어입니다.

import pandas as pd

data = {'삼성전자': [52200, 52300, 52900, 52000, 51700], 
           'LG전자': [68200, 67800, 68800, 67500, 66300]}
df = pd.DataFrame(data=data)
df.pct_change()
```

```python
# 하루 전이 아니라 2일 전에 사서 오늘 팔았을 때의 수익률을 계산할 수도 있습니다. 
# 이 경우 2일 전 데이터와 당일 데이터를 비교해야 합니다. 
# 다음 코드는 shift 메서드를 사용해서 수익률을 직접 계산합니다.

df / df.shift(2) - 1
```

```python
# pct_change 메서드의 periods 파라미터를 사용해서 2일 보유 수익률을 계산할 수 있습니다. 
# 검증된 메서드를 이용하면 의도치 않은 실수를 줄일 수 있어서 위와 같이 직접 구현하는 것보다는 
# pct_change 메서드를 사용하는 것이 좋습니다.

df.pct_change(periods=2)
```

```python
# 삼성전자와 LG전자의 데이터로 보유 후 2일 뒤 매도하는 전략의 누적 수익률을 계산해 봅시다. 
# pct_change로 인접한 두 값의 비율(수익률)을 얻어와서 원금 1을 더한 시리즈에 cumprod 메서드로 
# 누적수익률을 계산합니다. cumprod 메서드는 시리즈에 대해서 누적곱을 리턴합니다.

yeild = df.pct_change(periods=2) + 1
yeild.cumprod()
```

```python
# ss_ex_1.xlsx 엑셀 파일을 사용해서 각 분기의 시가 대비 일일 상승률(종가 기준)을 계산해 보겠습니다. 
# read_excel함수로 데이터를 읽어오며, usecols 파라미터로 일자, 시가, 종가 컬럼만을 선택합니다. 
# 날짜 컬럼을 데이터프레임의 인덱스로 지정하고 데이터는 일자를 기준으로 오름차순 정렬합니다.
# [ 그림 6.6.4 ]

import pandas as pd

df = pd.read_excel("data/ss_ex_1.xlsx", index_col=0, usecols=[0, 1, 4])
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df.head()
```

```python
# 코드를 실행한 결과 그림 6.6.4와 같은 127개의 시가와 종가를 확인할 수 있습니다. 
# 2021년 1분기의 시가는 2021-02-15의 83800입니다. 
# 1분기에 해당하는 1월부터 3월은 83800을 기준으로 상승률을 계산해야 합니다. 
# 뿐만 아니라 2분기, 3분기, 4분기 각각의 시가와 비교해서 종가가 얼마만큼 상승했는지를 계산해야 합니다.

# 시가 컬럼에 resample 메서드를 사용해서 분기(quarter) 별로 데이터를 샘플링합니다. 
# 이때 샘플링된 각 그룹의 첫 번째 값을 선택하도록 first 메서드를 적용합니다. 
# first 메서드를 호출한 결과는 시리즈인데 to_frame 메서드를 사용해서 데이터프레임으로 변경합니다.

# [ 그림 6.6.5 ]

df_quarter = df['시가'].resample('q').first().to_frame()
df_quarter
```

```python
# Grouper 객체를 사용해서 분기별 데이터를 집계할 수도 있습니다.
df_quarter = df['시가'].groupby(pd.Grouper(freq='q')).first().to_frame()
df_quarter
```

```python
# 수익률을 계산하기 위해 그림 6.6.4와 그림 6.6.5를 '분기별'로 연결해야 합니다. 
# 따라서 단순히 이어 붙이는 concat 보다는 merge를 사용해야 하며, 데이터프레임을 병합하기 위해 
# 기준이 되는 '분기' 컬럼을 각 데이터프레임에 추가해야 합니다.

df_quarter['quarter'] = df_quarter.index.quarter
df['quarter'] = df.index.quarter
```

```python
# 다음 코드는 필요한 데이터만을 슬라이싱해서 두 개의 데이터프레임을 병합합니다. 
# merge 함수는 기본적으로 인덱스를 제외한 값만을 합치기 때문에 reset_index 메서드로 인덱스를 값으로 치환한 뒤에 
# merge합니다.

df_daily = df[['종가', 'quarter']].reset_index()
r = pd.merge(left=df_daily, right=df_quarter, on='quarter')
```

```python
# 다음 코드는 수익률을 계산하고 그림 6.6.8과 같이 멀티 인덱스로 데이터를 정리합니다.

r['수익률'] = r['종가'] / r['시가']
r = r.set_index(['quarter','일자'])
r
```
