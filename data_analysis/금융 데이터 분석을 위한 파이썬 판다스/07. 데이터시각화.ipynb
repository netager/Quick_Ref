## 01) pyplot 기초

- 액자에 그려진 그림은 액자와 그림이 그려진 종이로 구성됩니다. 이와 유사하게 pyplot으로 그림을 그리기 위해서는 액자 역할을 하는 Figure 타입의 객체와 종이 역할을 하는 AxesSubplot의 객체가 필요합니다.
- Figure, Axes, Axis(x축), Axis(y축)


```python
# Matplotlib 라이브러리로 그래프를 그릴 때 데이터에 한글이 포함되어 있으면 에러가 발생합니다. 
# 이는 기본 설정된 폰트가 한글을 지원하지 않기 때문입니다. 이런 문제를 해결하기 위하여 다음과 같은 코드를 
# 사용하여 먼저 코드가 실행되는 환경이 macOS이면 'AppleGothic' 폰트로 설정하고 그렇지 않다면 한글 윈도우 
# 기본 폰트인 'Malgun Gothic'으로 설정합니다.

import matplotlib.pyplot as plt
import platform

%matplotlib inline

if platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic') 
else:
    plt.rc('font', family='Malgun Gothic')
```

```python
# 다음은 pyplot을 이용해서 그래프를 그리는 가장 기본적인 코드입니다.

import matplotlib.pyplot as plt 

fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1) # 1행 1열의 액자에 첫 번째 종이를 만들어라
                                   # 함수는 종이를 의미하는 AxesSubplot 객체를 반환함.
subplot.plot([1, 2, 3, 4]) # AxesSubplot 객체에 그래프를 그림
plt.show() # 액자를 보여줌
```

```python
# AxesSubplot은 그림이 그려질 수 있는 종이와 같은 역할을 합니다. 
# 한 액자에 두 개의 종이를 붙였다면 각 종이에 서로 다른 그림을 그릴 수 있는 것처럼 
# 여러 개의 AxesSubplot객체를 만든 후 각 객체에 플롯팅 할 수 있습니다.

import matplotlib.pyplot as plt 

fig = plt.figure()
subplot1 = fig.add_subplot(1, 2, 1)
subplot2 = fig.add_subplot(1, 2, 2)
subplot1.plot([1, 2, 3, 4])
subplot2.plot([4, 3, 2, 1])
plt.show()
```

```python
# Figure 객체와 AxesSubplot 객체를 생성하는 두 가지 작업을 한 번에 하고자 할 때는 
# plt 모듈의 subplots 함수를 사용합니다. 다음 코드를 실행하면 2행 2열 그리드 형태의 AxesSubplot 객체가 생성됩니다.

import matplotlib.pyplot as plt 

fig, axes = plt.subplots(2, 2)

axes[0][0].plot([1, 2, 3, 4])
axes[0][1].plot([4, 3, 2, 1])
axes[1][0].plot([1, 2, 2, 1])
axes[1][1].plot([2, 1, 1, 2])
plt.show()
```

```python
 # 간단히 그래프를 그려볼 때는 다음과 같이 plt 모듈의 plot 함수를 바로 사용할 수도 있습니다. 
 # 이 경우 여러분은 명시적으로 Figure나 AxesSubplot 객체를 생성하지 않고도 그림을 그릴 수 있습니다. 
 # pyplot 모듈의 plt 함수는 Figure와 AxesSubplot 객체가 없는 경우 이를 먼저 생성한 후 AxesSubplot에 그래프를 그려줍니다. 
 # 복잡한 설정 없이 간단한 그래프를 그릴 때 편리한 방법입니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4])
plt.show() 
```

## 02) 라인(Line)) 차트

```python
# 앞서 그렸던 그래프처럼 각 값을 선으로 연결한 그래프를 라인 차트라고 합니다. 
# plot 함수로 리스트를 전달하면 해당 값을 y 축으로 표시합니다. 
# x축은 데이터를 전달하지 않으면 0부터 시작하여 1씩 증가하는 정수 값을 사용합니다. 
# 다음 코드는 (x=0, y=2), (x=1, y=1), (x=2, y=1), (x=3, y=2)의 4개의 점을 선으로 연결

import matplotlib.pyplot as plt 

plt.plot([2, 1, 1, 2])
plt.show()
```

```python
# x축과 y축의 값을 명시적으로 전달하는 경우 두 값의 개수가 다르다면 당연히 에러가 발생합니다.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3]

plt.plot(x, y)
plt.show()
```

```python
# 차트를 그릴 때 그래프의 제목과 각 축의 의미를 표시하여 어떤 데이터를 시각화한 것인지 알려주는 것이 중요합니다. 
# pyplot 모듈의 xlabel, ylabel, title 함수를 사용해서 그림 7.2.2와 같이 각 축에 대한 레이블 및 타이틀을 
# 추가할 수 있습니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [2, 1, 1, 2])

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("matplotlib")
plt.show()
```

```python
# AxesSubplot 객체를 사용해서도 label과 title을 표시할 수도 있습니다.

import matplotlib.pyplot as plt 

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 1, 1, 2])

ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("title")

plt.show()
```

```python
# 하나의 AxesSubplot에 여러 그래프를 그려야 할 때도 있습니다. 이 경우 여러 번 plot 함수를 호출하면 됩니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.plot([1, 2, 3, 4], [4, 3, 2, 1])

plt.show()
```

```python
# 한 AxesSubplot에 plot 함수를 여러 번 호출하여 그래프를 그린 경우 각 그래프를 구분할 수 있도록 
# 레이블을 붙여주는 것이 좋습니다. plot 함수에서 label 파라미터로 사용할 레이블을 문자열로 전달하고 
# legend 함수를 호출합니다. 'best' 파라미터는 최적의 위치에 범례를 표시합니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='ascending')
plt.plot([1, 2, 3, 4], [4, 3, 2, 1], label='descending')

plt.legend(loc='best')
plt.show()
```

```python
# 이번에는 코스피 지수 데이터를 사용하여 2020년의 월별 평균 거래량을 막대 차트로 그려보겠습니다. 
# 엑셀 파일로부터 데이터를 로드해서 데이터프레임의 resample 메서드로 월별 평균을 계산합니다.

import pandas as pd

kospi = pd.read_excel("data/kospi.xlsx", index_col='일자')
#kospi.index = pd.to_datetime(kospi.index)
kospi2020 = kospi.loc['2020']
mean_kospi_2020 = kospi2020.resample('MS').mean()
mean_kospi_2020.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1,1,1)
month = [x.strftime("%Y-%m") for x in mean_kospi_2020.index]
ax.bar(x=month, height=mean_kospi_2020['거래량'],width=0.5, color='skyblue')

plt.ticklabel_format(axis='y', style='plain') # 지수로 표현하지 않음
plt.title("KOSPI 월별 평균 거래량 (2020년)")
plt.xlabel('월')
plt.ylabel('거래량')
plt.show()
```

## 04) 히스토그램

```python
# 히스토그램은 특정 변수에 대해 각 구간별로 나눈 후 구간 별 빈도수를 그린 그래프입니다. 
# pyplot 모듈의 hist 함수를 사용하면 히스토그램 그래프를 그릴 수 있습니다.

import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data)
plt.show()
```

```python
# hist 함수는 입력 데이터를 구간으로 나눈 후 각 구간의 빈도수를 계산합니다. 
# hist 함수의 bins 파라미터로 값을 전달하지 않으면 데이터를 10개의 구간으로 나눕니다. 
# 다음 코드에서 0~99의 100개의 정수로 히스토그램을 그리면 총 10개의 구간에 빈도수가 10이 됩니다.

import matplotlib.pyplot as plt

data = range(100) # 0 ~ 99
(n, bins, patches) = plt.hist(data)
plt.show()
```

```python
# hist 함수는 튜플 객체를 리턴합니다. 해당 튜플 객게는 각 구간별 빈도수(n), 구간(bins), 패치(patches) 정보를 포함
n, bins, patches
```

```python
# 위 그래프를 참고하면 삼성전자와 SK하이닉스의 가격 추이가 비슷해 보입니다. 
# 다만 종목 간에 가격 차이가 존재하므로 이를 정확히 비교하기 위해서는 두 종목을 같은 기준으로 
# 스케일을 조정해야 합니다. 이러한 과정을 지수화(indexing)라고 합니다.

# 일반적으로 지수화에서는 비교 구간의 첫 번째 값을 100으로 설정하고 나머지 값들의 스케일을 조정합니다. 
# 종가의 스케일을 조정한 후 다시 그래프를 그려보겠습니다. 
# 삼성전자의 종가 전체에 대해서 2020년 첫 번째 거래일로 나눈 후 100을 곱합니다. SK하이닉스의 종가에 대해서도 동일하게 적용합니다.

import matplotlib.pyplot as plt 

sec_index = (sec['Close'] / sec['Close'][0]) * 100
skhynix_index = (skhynix['Close'] / skhynix['Close'][0]) * 100

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(sec.index, sec_index, label="SEC")
ax.plot(sec.index, skhynix_index, label="SK hynix")
plt.ylabel("가격지수")
plt.title("삼성전자, SK하이닉스 2020년 가격 비교")
plt.legend()
plt.show()
```

```python
# 삼성전자의 일봉 데이터 60개를 사용하여 캔들 차트로 그려보겠습니다. 
# 캔들의 색상은 상승에 빨간색, 하락에 파란색으로 설정했습니다.

mc = mpf.make_marketcolors(
    up="r", 
    down="b", 
    edge="inherit",   # 캔들의 몸통색
    wick="inherit"    # 캔들의 머리/꼬리색
)

s = mpf.make_mpf_style(
    base_mpf_style="starsandstripes", 
    marketcolors=mc, 
    gridaxis='both',   # horizontal, vertical, both
    y_on_right=True    # False는 y축을 왼쪽에 표시
)

mpf.plot(data=df.iloc[:60], type='candle', style=s, figratio=(13, 6))
```

```python
# 이번에는 거래량을 같이 표시해 보겠습니다. 보통 거래량은 캔들 차트 아래에 표시합니다. 
# 거래량은 막대 차트 형태로 그려지는데 상승장의 거래량은 빨간색으로 하락장의 거래량은 파란색으로 표시되도록 설정했습니다. 
# plot 함수에서 volume 파라미터를 True로 설정하면 거래량이 같이 출력됩니다. 
# scale_width_adjustment 파라미터를 통해 일봉과 거래량 그래프의 너비를 조정할 수 있습니다. 
# 기본값은 1인데 이보다 더 크거나 작은 값을 설정할 수 있습니다.

mc = mpf.make_marketcolors(
    up="r", 
    down="b", 
    edge="inherit",   # 캔들의 몸통색
    wick="inherit",   # 캔들의 머리/꼬리색
    volume="inherit"  # 거래량 색상
)

s = mpf.make_mpf_style(
    base_mpf_style="starsandstripes", 
    marketcolors=mc, 
    gridaxis='both',   # horizontal, vertical, both
    y_on_right=True    # False는 y축을 왼쪽에 표시
)

mpf.plot(
    data=df.iloc[:60], 
    type='candle', 
    style=s, 
    figratio=(13, 6),
    volume=True,      # volume
    scale_width_adjustment=dict(volume=0.8, candle=1)
)
```

```python
# mplfinance를 사용하면 쉽게 이동평균선을 그릴 수 있습니다. plot 함수에서 mav 파라미터에 이동평균선의 기간을 튜플로 전달하면 됩니다.
# mav 파라미터로 (5, 20)의 튜플을 전달했습니다. 일봉 데이터를 기준으로 최근 5개와 20개 종가의 평균값이 라인 차트로 그려지는 것을 확인할 수 있습니다.

mpf.plot(
    data=df.iloc[:60], 
    type='candle',
    mav=(5, 20),   # moving average
    style=s, 
    figratio=(13, 6),
    volume=True,      # volume
    scale_width_adjustment=dict(volume=0.8, candle=1)
)
```

## 07) 이동평균선

```python
# 앞에서는 mplfinance 모듈을 사용해서 이동평균선을 플롯팅했습니다. 
# 이번에는 판다스 데이터프레임에서 제공하는 rolling 메서드를 사용해서 직접 이동평균 컬럼을 추가하고 
# 이를 matplotlib을 사용해서 플롯팅 해보겠습니다. 
# 먼저 엑셀 파일로부터 코스피 지수 데이터를 로드합니다. 'Close' 컬럼의 값을 사용하여 20, 60, 120일 이동평균값을 계산합니다.

import pandas as pd

kospi = pd.read_excel("data/kospi.xlsx", index_col=0)
kospi["MA20"] = kospi['종가'].rolling(window=20).mean()
kospi["MA60"] = kospi['종가'].rolling(window=60).mean()
kospi["MA120"] = kospi['종가'].rolling(window=120).mean()
kospi.head()
```

```python
# 종가 컬럼과 이동평균값 컬럼을 하나의 차트로 플롯팅해봅시다.

import matplotlib.pyplot as plt 
import platform

# 한글처리
if platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic') 
else:
    plt.rc('font', family='Malgun Gothic') 

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(kospi['종가'], label="close")  # 종가
ax.plot(kospi['MA20'], label="ma20")  # 20일 이동평균
ax.plot(kospi["MA60"], label="ma60")  # 60일 이동평균
ax.plot(kospi["MA120"], label="ma120")  # 120일 이동평균

ax.set_xlabel("날짜")
ax.set_ylabel("코스피지수")
ax.set_title("2019년 이후 코스피지수")
ax.legend(loc='best')
#plt.grid(True, axis='y')
plt.show()
```

```python
# matplotlib는 차트를 세밀하게 제어할 수 있기는 하지만 간단히 데이터의 추세를 확인하고 싶은 
# 목적 대비 타이핑해야 할 코드가 생각보다 많습니다. 데이터프레임에는 matplotlib를 기반으로하는 
# 시각화 메서드가 포함돼 있어서 빠르게 데이터의 특성을 확인할 수 있습니다. 
# 필요한 경우 matplotlib와 함께 사용할 수도 있습니다. 우선 실습할 데이터를 읽어와 봅시다. 
# 다음 코드는 코스피 지수의 가격 정보가 저장된 엑셀에서 세 개의 컬럼만을 얻어옵니다.

import pandas as pd

kospi = pd.read_excel("data/kospi.xlsx", index_col=0, usecols=[0, 1, 4])
kospi.sort_index(inplace=True)
kospi.head()
```

```python
# 시가 대비 종가의 상승/보합/하락 유무를 판별하고 집계해서 상승/보합/하락의 건수를 막대 차트로 그려보겠습니다. 
# 우선 조건 비교를 통해서 상승/보합/하락을 집계합니다. 
# groupby의 count 메서드를 사용해도 되지만 agg 메서드를 사용하면 지정한 컬럼만을 선택해서 함수를 적용하니 편리합니다.

upward    = kospi['시가'] < kospi['종가']
stationary  = kospi['시가'] == kospi['종가']
downward = kospi['시가'] > kospi['종가']            

kospi.loc[upward,   'category'] = "상승"
kospi.loc[stationary, 'category'] = "보합"
kospi.loc[downward, 'category'] = "하락"

how = {
    "category": len
}
data = kospi.groupby('category').agg(how)
print(data)
```
