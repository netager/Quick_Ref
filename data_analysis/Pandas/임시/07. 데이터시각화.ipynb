## 01) pyplot 기초

- 액자에 그려진 그림은 액자와 그림이 그려진 종이로 구성됩니다. 이와 유사하게 pyplot으로 그림을 그리기 위해서는 액자 역할을 하는 Figure 타입의 객체와 종이 역할을 하는 AxesSubplot의 객체가 필요합니다.
- Figure, Axes, Axis(x축), Axis(y축)


```python
# 다음은 pyplot을 이용해서 그래프를 그리는 가장 기본적인 코드입니다.

import matplotlib.pyplot as plt 

fig = plt.figure()
subplot = fig.add_subplot(1, 1, 1) # 1행 1열의 액자에 첫 번째 종이를 만들어라
                                   # 함수는 종이를 의미하는 AxesSubplot 객체를 반환함.
subplot.plot([1, 2, 3, 4]) # AxesSubplot 객체에 그래프를 그림
plt.show() # 액자를 보여줌
```

```python
# AxesSubplot은 그림이 그려질 수 있는 종이와 같은 역할을 합니다. 
# 한 액자에 두 개의 종이를 붙였다면 각 종이에 서로 다른 그림을 그릴 수 있는 것처럼 
# 여러 개의 AxesSubplot객체를 만든 후 각 객체에 플롯팅 할 수 있습니다.

import matplotlib.pyplot as plt 

fig = plt.figure()
subplot1 = fig.add_subplot(1, 2, 1)
subplot2 = fig.add_subplot(1, 2, 2)
subplot1.plot([1, 2, 3, 4])
subplot2.plot([4, 3, 2, 1])
plt.show()
```

```python
# Figure 객체와 AxesSubplot 객체를 생성하는 두 가지 작업을 한 번에 하고자 할 때는 
# plt 모듈의 subplots 함수를 사용합니다. 다음 코드를 실행하면 2행 2열 그리드 형태의 AxesSubplot 객체가 생성됩니다.

import matplotlib.pyplot as plt 

fig, axes = plt.subplots(2, 2)

axes[0][0].plot([1, 2, 3, 4])
axes[0][1].plot([4, 3, 2, 1])
axes[1][0].plot([1, 2, 2, 1])
axes[1][1].plot([2, 1, 1, 2])
plt.show()
```

```python
 # 간단히 그래프를 그려볼 때는 다음과 같이 plt 모듈의 plot 함수를 바로 사용할 수도 있습니다. 
 # 이 경우 여러분은 명시적으로 Figure나 AxesSubplot 객체를 생성하지 않고도 그림을 그릴 수 있습니다. 
 # pyplot 모듈의 plt 함수는 Figure와 AxesSubplot 객체가 없는 경우 이를 먼저 생성한 후 AxesSubplot에 그래프를 그려줍니다. 
 # 복잡한 설정 없이 간단한 그래프를 그릴 때 편리한 방법입니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4])
plt.show() 
```

## 02) 라인(Line)) 차트

```python
# 앞서 그렸던 그래프처럼 각 값을 선으로 연결한 그래프를 라인 차트라고 합니다. 
# plot 함수로 리스트를 전달하면 해당 값을 y 축으로 표시합니다. 
# x축은 데이터를 전달하지 않으면 0부터 시작하여 1씩 증가하는 정수 값을 사용합니다. 
# 다음 코드는 (x=0, y=2), (x=1, y=1), (x=2, y=1), (x=3, y=2)의 4개의 점을 선으로 연결

import matplotlib.pyplot as plt 

plt.plot([2, 1, 1, 2])
plt.show()
```

```python
# x축과 y축의 값을 명시적으로 전달하는 경우 두 값의 개수가 다르다면 당연히 에러가 발생합니다.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3]

plt.plot(x, y)
plt.show()
```

```python
# 차트를 그릴 때 그래프의 제목과 각 축의 의미를 표시하여 어떤 데이터를 시각화한 것인지 알려주는 것이 중요합니다. 
# pyplot 모듈의 xlabel, ylabel, title 함수를 사용해서 그림 7.2.2와 같이 각 축에 대한 레이블 및 타이틀을 
# 추가할 수 있습니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [2, 1, 1, 2])

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("matplotlib")
plt.show()
```

```python
# AxesSubplot 객체를 사용해서도 label과 title을 표시할 수도 있습니다.

import matplotlib.pyplot as plt 

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 1, 1, 2])

ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("title")

plt.show()
```

```python
# 하나의 AxesSubplot에 여러 그래프를 그려야 할 때도 있습니다. 이 경우 여러 번 plot 함수를 호출하면 됩니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.plot([1, 2, 3, 4], [4, 3, 2, 1])

plt.show()
```

```python
# 한 AxesSubplot에 plot 함수를 여러 번 호출하여 그래프를 그린 경우 각 그래프를 구분할 수 있도록 
# 레이블을 붙여주는 것이 좋습니다. plot 함수에서 label 파라미터로 사용할 레이블을 문자열로 전달하고 
# legend 함수를 호출합니다. 'best' 파라미터는 최적의 위치에 범례를 표시합니다.

import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='ascending')
plt.plot([1, 2, 3, 4], [4, 3, 2, 1], label='descending')

plt.legend(loc='best')
plt.show()
```

```python
# Matplotlib 라이브러리로 그래프를 그릴 때 데이터에 한글이 포함되어 있으면 에러가 발생합니다. 
# 이는 기본 설정된 폰트가 한글을 지원하지 않기 때문입니다. 이런 문제를 해결하기 위하여 다음과 같은 코드를 
# 사용하여 먼저 코드가 실행되는 환경이 macOS이면 'AppleGothic' 폰트로 설정하고 그렇지 않다면 한글 윈도우 
# 기본 폰트인 'Malgun Gothic'으로 설정합니다.

import matplotlib.pyplot as plt
import platform

if platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic') 
else:
    plt.rc('font', family='Malgun Gothic')
```

```python
# 이번에는 코스피 지수 데이터를 라인 차트로 그려보겠습니다. 
# 2019년부터 2021년 8월까지의 코스피 지수 데이터를 엑셀 파일로부터 데이터프레임으로 로드합니다.

import pandas as pd

kospi = pd.read_excel("data/kospi.xlsx", index_col="일자")

kospi.sort_index(inplace=True)
kospi
```

```python
# 로드된 코스피 데이터에서 '종가' 컬럼의 값을 라인 차트로 그려봅시다. 
# 이 경우 데이터프레임의 인덱스가 자동으로 x축으로 사용됩니다. 
# Figure 클래스의 객체를 생성할 때 크기를 설정했으며 pyplot의 grid 함수를 사용하여 y축 방향으로 그리드를 표시했습니다.

import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(kospi['종가'])

ax.set_xlabel("날짜")
ax.set_ylabel("코스피지수")
ax.set_title("2019년 이후 코스피지수")
plt.grid(True, axis='y')
plt.show()
```

```python
# 다음 코드는 read_excel 함수의 index_col 파라미터로 첫 번째 컬럼을 인덱스로 지정합니다. 
# header 파라미터에는 [0, 1]을 사용해서 두 개의 행으로 멀티 컬럼을 구성합니다.

df = pd.read_excel("data/loan.xlsx", index_col=0, header=[0, 1])
df.head()
```

```python
# 대출잔액 부분만 그래프로 시각화해 봅시다. 
# 데이터프레임에서 대출잔액을 인덱싱한 후 과거 날짜가 위로 오도록 정렬합니다.

대출잔액 = df['대출잔액']
대출잔액 = 대출잔액[::-1]
대출잔액.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 8)) 
plt.plot(대출잔액.index, 대출잔액["전체"], marker='o', label="전체")
plt.plot(대출잔액.index, 대출잔액["1금융권"], marker='o', label="1금융권") 
plt.plot(대출잔액.index, 대출잔액["2금융권"], marker='o', label="2금융권")
plt.legend(loc='best')
plt.rc("axes", unicode_minus=False) # y축 음수처리 plt.ylabel("(전년동월비,%)") plt.legend() plt.grid(True, axis='y') plt.show()
```
