#1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
#matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df_m = pd.read_excel('202208_202208_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.head()
```

```python
# 숫자 타입의 , 제거
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('202208_202208_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head()
```

```python
df_m.columns
```

```python
df_w.columns = df_m.columns
```

```python
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w.iloc[0]
```

## 데이터 시각화

```python
plt.figure(figsize=(10,8))
plt.title('2022년 8월 인구수')
plt.barh(df_m.columns, -df_m.iloc[0] // 1000, label='남자')
plt.barh(df_w.columns, df_w.iloc[0] // 1000, label='여자')
plt.savefig('2022_인구피라미드.png', dpi=100)
plt.show()
```

```python
df_m_12 = pd.read_excel('201208_201208_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m_12.iloc[0] = df_m_12.iloc[0].str.replace(',','').astype(int)

df_w_12 = pd.read_excel('201208_201208_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w_12.iloc[0] = df_w_12.iloc[0].str.replace(',','').astype(int)

df_w_12.columns = df_m_12.columns

plt.figure(figsize=(10,8))
plt.title('2012년 8월 인구수')
plt.barh(df_m_12.columns, -df_m_12.iloc[0] // 1000, label='남자')
plt.barh(df_w_12.columns,  df_w_12.iloc[0] // 1000, label='여자')
plt.savefig('2012_인구피라미드.png', dpi=100)
plt.show()
```
