# 원 그래프 (기본)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
#matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30,25,20,13,10,2]
labels=['Python','Java','Javascript','C#','C/C++', 'ETC']
```

```python
plt.figure(figsize=(10,5))
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
plt.figure(figsize=(10,5))
explode=[0.2,0.1,0,0,0,0]
plt.pie(values, labels=labels, autopct='%.1f%%', explode=explode)
plt.show()


```

```python
plt.figure(figsize=(10,5))
explode=[0.05] * 6
plt.pie(values, labels=labels, autopct='%.1f%%', explode=explode)
plt.show()
```

```python
plt.figure(figsize=(10,5))
explode=[0.05] * 6
plt.pie(values, labels=labels, autopct='%.1f%%', explode=explode)
plt.legend(loc=[1.2, 0.3])

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('언어별 선호도')
explode=[0.05] * 6
plt.pie(values, labels=labels, autopct='%.1f%%', explode=explode)
plt.legend(loc=[1.2, 0.3], title='언어별 선호도')

plt.show()
```
