#16. 여러 그래프

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
#matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python

fig, axes = plt.subplots(2, 2, figsize=(15,10))
fig.suptitle('여러 그래프')

# 첫 번째 그래프
axes[0, 0].bar(df['이름'], df['국어'], label='국어점수')
axes[0, 0].set_title('첫번째 그래프')
axes[0, 0].legend()
axes[0, 0].set(xlabel='이름', ylabel='점수')
axes[0, 0].set_facecolor('lightyellow')
axes[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axes[0, 1].plot(df['이름'], df['수학'], label='수학')
axes[0, 1].plot(df['이름'], df['영어'], label='수학')
axes[0, 1].set_title('두번째 그래프')
axes[0, 1].legend()

# 세 번째 그래프
axes[1, 0].barh(df['이름'], df['키'], label='키')

# 네 번째 그래프
axes[1, 1].plot(df['이름'], df['사회'], color='green', alpha=0.5)
```
