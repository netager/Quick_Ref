#8. 막대 그래프 (기본)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
#matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python

df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
plt.bar(df['이름'], df['키'])
```

```python
plt.bar(df['이름'], df['키'], color='r')
```

```python
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors)
```

```python
plt.figure(figsize=(10,5))
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors, alpha=0.5)
```

```python
plt.figure(figsize=(10,5))
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors, alpha=0.5)
plt.ylim(175, 195)
```

```python
plt.figure(figsize=(10,5))
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors, alpha=0.5, width=0.5)
```

```python
plt.figure(figsize=(10,5))
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors, alpha=0.5, width=0.5)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
ticks=['1번학생','2번학생','3번학생','4번학생','5번학생','6번학생','7번학생','8번학생']

plt.figure(figsize=(10,5))
colors = ['r','b','g','r','b','g','r','b']
plt.bar(df['이름'], df['키'], color=colors, alpha=0.5, width=0.5)
plt.yticks(rotation=45)
plt.xticks(df['이름'], ticks, rotation=45)
```
