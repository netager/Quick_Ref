# 4. 스타일

```python
import matplotlib.pyplot as plt

# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac

# 글자 크기
matplotlib.rcParams['font.size'] = 15 # 글자 크기

# 한글 폰트 사용시 마이너스 글자가 깨지는 현상 해결
matplotlib.rcParams['axes.unicode_minus'] = False

# 그래프 크기
plt.figure(figsize=(10, 5))

%matplotlib inline
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

# 마커

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='^', markersize=10)
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='D', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=1)
```

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.2)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.2)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=200) # dots per inch, 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#85b0ed')
plt.plot(x, y)
```
