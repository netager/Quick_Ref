# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt

# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac

# 글자 크기
matplotlib.rcParams['font.size'] = 15 # 글자 크기

# 한글 폰트 사용시 마이너스 글자가 깨지는 현상 해결
matplotlib.rcParams['axes.unicode_minus'] = False

# 그래프 크기
plt.figure(figsize=(10, 5))

%matplotlib inline
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # 시계 방향으로 
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
#explode = [0.1, 0.15, 0, 0, 0, 0]
explode = [0.05] * 6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도', color='w')
plt.legend(loc=(1.2, 0.3), title='언어')
```
