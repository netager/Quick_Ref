# 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt

# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac

# 글자 크기
matplotlib.rcParams['font.size'] = 15 # 글자 크기

# 한글 폰트 사용시 마이너스 글자가 깨지는 현상 해결
matplotlib.rcParams['axes.unicode_minus'] = False

# 그래프 크기
plt.figure(figsize=(10, 5))

%matplotlib inline
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]
colors = ['r','g','b']
plt.bar(labels, values, color=colors)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]
colors = ['r','g','b']
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]
colors = ['r','g','b']

plt.bar(labels, values, color=colors, alpha=0.5)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # X 축의 이름 데이터 각도를 45 도로 설정
plt.yticks(rotation=45) # Y 축의 이름 데이터 각도를 45 도로 설정
```

```python
labels = ['강백호','서태웅','정대만']
values = [190, 187, 184]
ticks = ['1번학생','2번학생','3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
plt.show()
```
